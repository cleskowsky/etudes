```python
def opposite(x, y):
    "Return true if x is the capital char version of y or vice versa"
    return y == x.upper() or x == y.upper()

print(
    opposite('a', 'A'),
    opposite('A', 'a'),
    opposite('a', 'b')
)
```

```python
def react(p):
    "React p once and return a new polymer"
    for i in range(len(p) - 1):
        j = i + 1
        if opposite(p[i], p[j]):
            return p[:i] + p[j+1:], True
    return p, False

print(react("bAa"), react('aA'), react('aAb'), react('aBbA'))
```

```python
def reduce(p):
    while True:
        p, reacted = react(p)
        if not reacted: return p

p = reduce(data)
print(len(p), p)
```

```python
# from Peter Norvig's pytudes
import re

input5 = data

alphabet = 'abcdefghijklmnopqrstuvwxyz'
ALPHABET = alphabet.upper()

OR = '|'.join
reaction = OR(OR([c + C, C + c]) for c, C in zip(alphabet, ALPHABET))

def reduction(text):
    "Remove all cC or Cc pairs, and repeat."
    while True:
        text, n = re.subn(reaction, '', text)
        if n == 0:
            return text
        
reduction('dabAcCaCBAcCcaDA')
```
