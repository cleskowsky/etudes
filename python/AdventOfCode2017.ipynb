```python
from collections import defaultdict

def input(fname):
    """Reads instructions from input file"""
    with open(fname) as f:
        return tuple(map(vector, f.readlines()))

def vector(line):
    """Returns a vector of instruction and arguments"""
    return tuple(map(atom, line.split()))

def atom(token):
    """Returns t type-converted"""
    try:
        return int(token)
    except:
        # not a number
        return token

def lookup(regs, k):
    return (k if isinstance(k, int) else regs[k])

if __name__ == '__main__':
    program = input('day23.txt')
    regs = defaultdict(int)
    pc = 0

    while True:
        instr = program[pc]
        op, x, y = instr[0], instr[1], instr[2]
        vy = lookup(regs, y)
        if   op == 'set': regs[x] = vy
        elif op == 'sub': regs[x] -= vy
        elif op == 'mul': regs[x] *= vy
        elif op == 'jnz':
            if lookup(regs, x) != 0: pc += y
            next
        pc += 1
        # print(pc, regs)

        if pc > len(program) or pc < 0:
            break
```
